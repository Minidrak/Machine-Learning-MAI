10. Какой эквивалент enumerate для numpy массивов?

````python
# Есть np.ndenumerate()

array = np.random.randint(1,11,size=5)
print(array)

for ind, val in np.ndenumerate(array):
    print(ind, val)

print(array)

for ind, val in enumerate(array):
    print(ind, val)

print(array)
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
#!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt

df['timestamp'] = pd.to_datetime(df['timestamp']) # преобразования в формат дат

nagatinski_zaton = df[df['sub_area'] == 'Nagatinskij Zaton'] #была опечатка в задании в слове Nagatinski, в датасете есть только Nagatinskij Zaton

nagatinski_zaton_avg_price = nagatinski_zaton.resample('ME', on='timestamp')['price_doc'].mean().reset_index()
nagatinski_zaton_avg_price['price_doc'] = nagatinski_zaton_avg_price['price_doc'] / 1e6
sns.set(style="dark", palette="muted")

plt.figure(figsize=(12, 6))
sns.lineplot(x='timestamp', y='price_doc', data=nagatinski_zaton_avg_price, marker='o', linewidth=2)

plt.title('Изменение средней цены квартиры в Нагатинском Затоне(в млн рублей) с течением времени', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Средняя цена квартиры', fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()
````
